## Setup

```python
import openai
import urllib
import matplotlib.pyplot as plt

from keys import OPENAI_API_KEY  # indirect import of the API key

openai.api_key = OPENAI_API_KEY
```

```python
def get_generations(prompt, n=1):
    """
    Allow you to edit and extend an image by uploading a mask.
    """
    try:
        response = openai.Image.create(
            prompt=prompt,
            n=1,
            size='1024x1024',
        )
        return response['data'][0]['url']
    except openai.error.OpenAIError as error:
        print(error.http_status)
        print(error.error)


def show_image(image_url):
    """
    Show the generated image.
    """
    response = urllib.request.urlretrieve(image_url, 'generation.jpg')
    image = plt.imread(response[0])
    plt.imshow(image)
    plt.title('GENERATION')
    plt.axis('off')
    
    plt.show()
```

## Generations

```python
prompt = input(r"""""")
```

```python
image_url = get_generations(prompt)
show_image(image_url)
```
